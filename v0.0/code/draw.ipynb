```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw(filesrc):
    with open(filesrc, 'r') as file:
        lines = file.readlines()
        it = int(lines[0])
        L = float(lines[1])
        vt = float(lines[2])
        vb = int(lines[3])
        dt = float(lines[4])
        xp = np.array([float(x) for x in lines[5].split()])
        vp = np.array([float(x) for x in lines[6].split()])

    odd_indices = np.arange(0, len(xp), 2)
    even_indices = np.arange(1, len(xp), 2)
    plt.plot(xp[odd_indices], vp[odd_indices], 'r.', markersize=2)
    plt.plot(xp[even_indices], vp[even_indices], 'b.', markersize=2)

    plt.axis([0, L, -3 * (abs(vt) + abs(vb)), 3 * (abs(vt) + abs(vb))])
    plt.title(f'Phase space plotting, t={int(it * dt)}')
    plt.xlabel('xp')
    plt.ylabel('vp')
    plt.show()
```

```python
for i in range(4):
    draw('./output/'+str(i)+'.txt')
```

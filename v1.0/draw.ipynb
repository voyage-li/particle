```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = []
y = []
z = []
with open('result.txt') as f:
    lines = f.readlines()
    for it in lines:
        tmp = [float(x) for x in it.split()]
        x.append(tmp[0])
        y.append(tmp[1])
        z.append(tmp[2]) 

```

```python
for i in x:
    if abs(i) > 20:
        print(i)
```

```python
r = [np.sqrt(x[i]*x[i]+y[i]*y[i]) for i in range(len(x))]
plt.plot(r,z,'.',markersize=2)
plt.xlabel('sqrt(x^2+y^2)')
plt.ylabel('z')
plt.show()
```

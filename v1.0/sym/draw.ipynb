## 谐振子 辛算法

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('result.txt') as f:
    lines = f.readlines()
    q = np.array([float(x) for x in lines])

t = np.arange(0,1000*0.05,0.05)
plt.plot(t,q)
plt.axis([0,50,-3,3])
plt.xlabel('t')
plt.ylabel('q')
plt.show()
```
